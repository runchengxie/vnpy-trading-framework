## 3. 核心模块功能展示

### 3.1 数据处理 (`data_utils.py`)

这个模块负责从Alpaca获取历史数据、进行缓存、数据清洗、重采样，并添加技术指标和卡尔曼滤波平滑。

```python
# 导入数据处理工具和Alpaca API
from patf_trading_framework.data_utils import fetch_historical_data, apply_kalman_filter, add_technical_indicators
from alpaca_trade_api.rest import REST, TimeFrame

# 初始化API
API_KEY = os.getenv('APCA_API_KEY_ID')
SECRET_KEY = os.getenv('APCA_API_SECRET_KEY')
BASE_URL = os.getenv('ALPACA_BASE_URL', 'https://paper-api.alpaca.markets')

api = REST(API_KEY, SECRET_KEY, base_url=BASE_URL)
print("Alpaca API 初始化完成。")

# 获取数据配置
data_config = config['data']

# 使用 data_utils 获取数据
# 注意：我们直接传入配置，而不是硬编码
df_raw = fetch_historical_data(
    api=api,
    symbol=data_config['ticker'],
    timeframe=TimeFrame.Minute, # 先获取分钟数据再重采样
    start_date=data_config['start_date'],
    end_date=data_config['end_date'],
    cache_dir=config['paths']['cache_dir']
)

if df_raw is not None:
    print("原始分钟数据获取成功！")
    display(df_raw.head())

    # 重采样到15分钟
    resample_freq = data_config['resample_freq_template'].format(timeframe_value=data_config['timeframe_value'])
    agg_dict = {'open': 'first', 'high': 'max', 'low': 'min', 'close': 'last', 'volume': 'sum'}
    df = df_raw.resample(resample_freq).agg(agg_dict).dropna()
    
    # 添加技术指标和卡尔曼滤波
    df['filtered_close'] = apply_kalman_filter(df['close'])
    df = add_technical_indicators(df)
    
    print("\n数据处理和指标计算完成！")
    display(df[['close', 'filtered_close', 'ema_12', 'adx_14']].tail())

    # 可视化展示
    df[['close', 'filtered_close', 'sma_20', 'sma_50']].plot(figsize=(15, 7), title=f"{data_config['ticker']} 收盘价与指标")
    plt.show()
```

### 3.2 策略定义 (`strategies.py`)

此模块包含了所有交易策略的Backtrader实现。我们来看一下`MeanReversionZScoreStrategy`的核心逻辑。

- **信号生成**: 当价格的Z-Score低于预设的下轨 (`zscore_lower`) 时买入，高于上轨 (`zscore_upper`) 时卖出。
- **平仓**: 当Z-Score回归到 `exit_threshold` (通常是0附近) 时平仓。
- **数据源**: 策略可以选择使用原始收盘价或经过卡尔曼滤波平滑后的价格。

## 4. 策略回测与优化

现在，我们将所有组件整合起来，使用`backtrader`引擎对我们的策略进行回测。

### 4.1 单次运行回测 (Mean Reversion Strategy)

我们将对Z-Score均值回归策略进行一次标准的回测。

```python
import backtrader as bt
from patf_trading_framework.strategies import MeanReversionZScoreStrategy, PandasDataFiltered
from patf_trading_framework.backtest_utils import analyze_optimization_results

# 获取回测配置
backtest_config = config['backtest']
strategy_config = config['strategies']['mean_reversion']

# 准备数据feed
# 添加 openinterest 列，backtrader 需要它
df['openinterest'] = 0
data_feed = PandasDataFiltered(dataname=df)

# 创建 Cerebro 实例
cerebro = bt.Cerebro()
cerebro.adddata(data_feed)
cerebro.broker.setcash(backtest_config['initial_cash'])
cerebro.broker.setcommission(commission=backtest_config['commission'])

# 添加分析器
cerebro.addanalyzer(bt.analyzers.TradeAnalyzer, _name='tradeanalyzer')
cerebro.addanalyzer(bt.analyzers.SharpeRatio, _name='sharpe')
cerebro.addanalyzer(bt.analyzers.DrawDown, _name='drawdown')
cerebro.addanalyzer(bt.analyzers.Returns, _name.name='returns')

# 添加策略
cerebro.addstrategy(MeanReversionZScoreStrategy, **strategy_config['params'])

# 运行回测
print("开始运行单次回测...")
results = cerebro.run()
strat = results[0]
print("回测完成！")

# 打印结果
final_value = cerebro.broker.getvalue()
pnl = final_value - backtest_config['initial_cash']
print(f"\n--- 回测结果 ---")
print(f"初始资金: ${backtest_config['initial_cash']:,.2f}")
print(f"最终资金: ${final_value:,.2f}")
print(f"净盈亏:    ${pnl:,.2f}")

# 可视化结果
print("\n生成回测图表...")
# cerebro.plot() 会在notebook中直接生成一个交互式图表，或者一个图片
# 如果在非交互式环境，需要保存到文件
fig = cerebro.plot(style='candlestick', barup='green', bardown='red', returnfig=True)[0][0]
# 可以在这里保存 fig.savefig(...)
plt.show() # 在Notebook中显示
```

### 4.2 参数优化

找到最佳参数是策略成功的关键。下面我们来优化`MeanReversionZScoreStrategy`的`zscore_period`参数。

```python
# 创建新的Cerebro实例进行优化
cerebro_opt = bt.Cerebro()
cerebro_opt.adddata(data_feed)
cerebro_opt.broker.setcash(backtest_config['initial_cash'])
cerebro_opt.broker.setcommission(commission=backtest_config['commission'])

# 添加分析器
cerebro_opt.addanalyzer(bt.analyzers.TradeAnalyzer, _name='tradeanalyzer')
cerebro_opt.addanalyzer(bt.analyzers.SharpeRatio, _name='sharpe')
cerebro_opt.addanalyzer(bt.analyzers.DrawDown, _name='drawdown')
cerebro_opt.addanalyzer(bt.analyzers.Returns, _name='returns')

# 添加优化策略
opt_ranges = strategy_config['opt_ranges']
# 将列表转为可迭代对象
opt_params_ready = {key: range(*val) if len(val) == 2 else val for key, val in opt_ranges.items()}
cerebro_opt.optstrategy(MeanReversionZScoreStrategy, **opt_params_ready)

print("开始参数优化...")
optimized_results = cerebro_opt.run(maxcpus=1) # 在Notebook中建议用单核，避免复杂问题
print("优化完成！")

# 分析优化结果
param_names = list(opt_ranges.keys())
results_df = analyze_optimization_results(optimized_results, param_names)

print("\n--- 优化结果分析 (按最终市值排序) ---")
display(results_df.sort_values(by='Final Value', ascending=False).head(10))

print("\n--- 优化结果分析 (按夏普比率排序) ---")
display(results_df.sort_values(by='Sharpe Ratio', ascending=False).head(10))
```

## 5. 高级功能与实盘准备

除了回测，本框架还提供了风险管理、性能分析和API交互等高级功能。

### 5.1 风险管理 (`risk_manager.py`)

`RiskManager` 模块可以计算滚动VaR、检查流动性等。

```python
from patf_trading_framework.risk_manager import RiskManager

risk_manager = RiskManager()

# 用历史收益率数据填充风险管理器
for r in df['close'].pct_change().dropna():
    risk_manager.returns_history.append(r)

# 假设当前组合价值为100,000美元
portfolio_value = 100000 
var_result = risk_manager.calculate_var(portfolio_value, method='historical')

print("--- VaR 计算示例 ---")
print(f"置信度: {var_result['confidence']:.1%}")
print(f"计算方法: {var_result['method']}")
print(f"日VaR金额: ${var_result['var']:,.2f}")
print(f"日VaR百分比: {var_result['var_percentage']:.2%}")
```

### 5.2 券商API交互 (`broker_handler.py`)

`BrokerAPIHandler` 封装了与Alpaca API的所有交互，如获取账户信息、下单、查询订单等。

```python
from patf_trading_framework.broker_handler import BrokerAPIHandler

# 注意：这里的所有操作都是在Alpaca的Paper Trading（模拟交易）账户上进行
broker = BrokerAPIHandler()

# 获取账户信息
account_info = broker.get_account_info()
if account_info:
    print("--- 账户信息 ---")
    print(f"账户ID: {account_info.id}")
    print(f"状态: {account_info.status}")
    print(f"资产净值: ${float(account_info.portfolio_value):,.2f}")
    print(f"购买力: ${float(account_info.buying_power):,.2f}")
else:
    print("获取账户信息失败。")

# 查询当前持仓
positions = broker.list_positions()
if positions:
    print("\n--- 当前持仓 ---")
    for p in positions:
        print(f"- {p.symbol}: 数量={p.qty}, 市值=${float(p.market_value):,.2f}")
else:
    print("\n当前无持仓。")
```
